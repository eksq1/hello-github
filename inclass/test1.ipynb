## **2023년의 월별 강수량 데이터 분석**

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

lifesat = pd.read_csv("data3.csv")
x = lifesat[["month"]].values
y = lifesat[["mm"]].values

lifesat.plot(kind="scatter", grid=True, x="month", y="mm")
plt.axis([0, 12, 0, 400])
plt.show()

model = LinearRegression()

model.fit(x, y)

X_new = [[1]]
print(model.predict(X_new))
```

#### 위 데이터를 분석하여 2024년 1월 강수량은 73.1mm로 예측되었고, 실제 2024년 1월 강수량은 16.3mm였다.
