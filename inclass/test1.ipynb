```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

lifesat = pd.read_csv("month.csv")
x = lifesat[["month"]].values
y = lifesat[["mm"]].values

lifesat.plot(kind="scatter", grid=True, x="month", y="mm")
plt.axis([1, 12, 0, 400])
plt.show()

model = LinearRegression()

model.fit(x, y)

X_new = [[12]]
print(model.predict(X_new))
```
